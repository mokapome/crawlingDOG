```python
import os
import cv2

image_dir = "./dog_to_human/human"

# 이미지 파일 목록 가져오기
image_files = [f for f in os.listdir(image_dir) if f.endswith(".jpg")]

# 이미지를 크기 변환 덮어쓰기
for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)

    # 이미지 로드
    img = cv2.imread(image_path)

    # 이미지 크기를 64x64로 조정
    img_64x64 = cv2.resize(img, (64, 64))

    # 64x64 이미지를 원본 이미지로 덮어쓰기
    cv2.imwrite(image_path, img_64x64)

    print(f"{image_file} 이미지를 64x64 크기로 덮어썼습니다.")

print("덮어쓰기 완료.")
```

```python
import os
import time
import urllib.request
import cv2
import dlib
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

# 검색어 설정
search_query = "Kpop가수"

# Chrome 웹 드라이버 서비스 객체 생성
chrome_service = Service('C:\\chromedriver\\chromedriver.exe')  # 웹 드라이버 경로 지정

# Chrome 웹 드라이버 설정
driver = webdriver.Chrome(service=chrome_service)

# 구글 이미지 검색 페이지로 이동
driver.get("https://www.google.com/imghp?hl=ko&tab=wi&authuser=0&ogbl")

# 검색어 입력
elem = driver.find_element(By.NAME, "q")
elem.send_keys(search_query)
elem.send_keys(Keys.RETURN)

# 이미지 다운로드 경로 설정
download_path = "./people"

# JSON 파일 저장 경로 설정
json_path = os.path.join(download_path, "json")
os.makedirs(json_path, exist_ok=True)

# 이미지 다운로드 디렉토리 생성
os.makedirs(download_path, exist_ok=True)

# 스크롤해서 더 많은 이미지 로드
SCROLL_PAUSE_TIME = 0.1
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(SCROLL_PAUSE_TIME)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            break
    last_height = new_height

# 원하는 저장 개수 설정
desired_save_count = 150
saved_count = 1

# 이미지 다운로드 및 얼굴 영역 자르고 저장
image_elements = driver.find_elements(By.CSS_SELECTOR, ".rg_i.Q4LuWd")

# 얼굴 감지기 초기화 (Dlib 사용)
face_detector = dlib.get_frontal_face_detector()

for image_element in image_elements:
    try:
        # 이미지 클릭 및 로드 대기
        image_element.click()
        time.sleep(2)

        # 이미지 다운로드 경로 설정
        image_path = os.path.join(download_path, f"person_{saved_count}_face.jpg")

        # 이미지 다운로드
        urllib.request.urlretrieve(image_element.get_attribute('src'), image_path)

        # 이미지 로드 및 크기 조정 (256x256)
        img = cv2.imread(image_path)
        img = cv2.resize(img, (256, 256))

        # 이미지 그레이 스케일로 변환 얼굴 감지
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_detector(gray)

        # 얼굴이 감지된 경우에만 처리
        if len(faces) > 0:
            face = faces[0]  # 첫 번째 얼굴 정보만 사용

            x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()

            # 얼굴 영역을 자르고 마진 추가
            margin = 20
            face_cropped = img[max(0, y1 - margin):min(y2 + margin, img.shape[0]),
                               max(0, x1 - margin):min(x2 + margin, img.shape[1])]

            # 얼굴 이미지 저장
            cv2.imwrite(image_path, face_cropped)

            print(f"얼굴 이미지 {saved_count}_face 저장 완료.")

            saved_count += 1  # 저장된 이미지 개수 증가

        else:
            print(f"얼굴을 찾지 못했습니다: {image_path}")
            # 이미지 저장에 실패한 경우 파일 삭제
            os.remove(image_path)

        # 원하는 개수만큼 이미지를 저장한 경우 종료
        if saved_count >= desired_save_count:
            break
    except Exception as e:
        print(f"다운로드 실패: {str(e)}")

# 브라우저 종료
driver.quit()
```
